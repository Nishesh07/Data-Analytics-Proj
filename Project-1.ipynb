```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data
df_movies = pd.read_csv('movies.csv')
```

```python
#first few rows of data
df_movies.head()
```

```python
#checking for missing values
df_movies.isnull().sum()
```

```python
#removing missing values
df_movies = df_movies.dropna()
```

```python
#checking for null values after removing
df_movies.isnull().sum()
```

```python
#checking data types
df_movies.dtypes
```

```python
#changing the budget to integer from float
df_movies['budget'] = df_movies['budget'].astype('int64')
```

```python
df_movies.dtypes
```

```python
df_movies.head()
```

```python
#checking or the top 10 highest rated movies
high_rated = df_movies.sort_values('score',ascending = False).head(10)
```

```python
high_rated
```

```python
#checking or the top 10 lowest rated movies
low_rated = df_movies.sort_values('score',ascending = True).head(10)
```

```python
low_rated
```

```python
#checking the highest grossed movies
highest_gross = df_movies.sort_values('gross',ascending = False)
```

```python
highest_gross
```

```python
#check for any duplicates
duplicateRows = df_movies[df_movies.duplicated()]
```

```python
duplicateRows
```

```python
#checking for correlation between budget and gross revenue
#lets create a scatter plot
```

```python
sns.regplot(x = 'budget', y = 'gross',data = df_movies,line_kws ={'color':'red'}).set(title='Budget vs Gross_value')
```

```python
#we can see budget and gross value have a strong correlation
df_movies.corr()
```

```python
correlation_matrix = df_movies.corr()
```

```python
#creating a correlation heatmap for easier visualization
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
```
